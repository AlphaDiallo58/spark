#### **Qu'est-ce que PySpark ?**

> PySpark est l'API Python pour Apache Spark. Apache Spark est un framework open-source pour le traitement des données distribuées conçu pour être rapide. Il offre des interfaces pour les langages de programmation comme Python, Scala, Java et R. Spark peut traiter des données en mémoire, ce qui le rend beaucoup plus rapide que les systèmes traditionnels de traitement de données comme Hadoop.

#### **Caractéristiques non exhaustives de pySpark**
  1. **Traitement en mémoire** : Spark peut traiter les données en mémoire, ce qui accélère les calculs.

  2. **API** : PySpark offre diverses API pour le traitement de données structurées (DataFrames), les opérations SQL, l'apprentissage automatique (MLlib) et le traitement des flux (Structured Streaming).

  3. **Tolérance aux Pannes** : Spark assure la tolérance aux pannes grâce à son modèle de calcul basé sur les RDD (Resilient Distributed Datasets) et les DataFrames.

  4. **Évolutivité** : Spark peut évoluer des petits clusters locaux à des clusters de grande échelle avec des milliers de nœuds.

#### **Comparaison avec Hadoop MapReduce**

> Hadoop MapReduce est un modèle de programmation pour le traitement de grandes quantités de données de manière distribuée. Cependant, il présente plusieurs limitations par rapport à Spark :

  - **Performance** : Spark est plus rapide que Hadoop MapReduce pour certaines charges de travail, grâce à son traitement en mémoire.

  - **Facilité d'utilisation** : PySpark offre des API plus simples et plus expressives que le modèle MapReduce. Après il est important de rappeler que la notion de simplicité ou facilté est relative et dépends des utilisateurs et de leur backgroud. ### **Prétaitement et tokénisation**

> Maintenant que nous avons compris la distribution de nos données, ayons pris connaissance que nos données contiennent des caractères qui ne nous seront pas utiles pour la suite de l'analyse. Nous avons également visualisé les mots les plus fréquents dans les deux classes. Nous pouvons maintenant passer au prétraitement des données et à la tokénisation.

> Malgré qu'il existe des bibliothèques de prétaitement de données, la plus part sont plus optimisées pour des données en anglais. Nous décidons donc de créer notre propre fonction de nettoyage.

> Nous allons créer une classe qui aura une méthode qui nous permettra de supprimer les caractères numériques, les liens, les ponctuations et tout ce qui nous semble non utile. Ensuite une autre méthode nous permettra de tokeniser, c'est à dire decouper en petites textes et de suprrimer les stop word ( comme 'le','la', 'mais'...)

```python
class CustomTextPreprocessor:
    def __init__(self):
        self.stop_words = set(stopwords.words('french'))

    def clean_text(self, text):
        text = text.lower()
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
        text = re.sub(r'\@\w+|\#','', text)
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'\d+', '', text)
        return text

    def tokenize_and_remove_stopwords(self, text):

        tokens = word_tokenize(text)
        tokens = [word for word in tokens if word not in self.stop_words]
        return tokens

    def preprocess(self, text):
        cleaned_text = self.clean_text(text)
        tokens = self.tokenize_and_remove_stopwords(cleaned_text)
        return tokens

    def preprocess_dataframe(self, df, input_col, output_col):
        preprocess_udf = udf(lambda x: self.preprocess(x), ArrayType(StringType()))
        return df.withColumn(output_col, preprocess_udf(df[input_col]))

# Crée une instance de CustomTextPreprocessor
custom_preprocessor = CustomTextPreprocessor()
```

```python
# Prétraiter les données
train_processed = custom_preprocessor.preprocess_dataframe(train_df, "message", "processed_message")
test_processed = custom_preprocessor.preprocess_dataframe(test_df, "message", "processed_message")
noclass_processed = custom_preprocessor.preprocess_dataframe(noclass_df, "message", "processed_message")

# Afficher les résultats pour vérifier le prétraitement
train_processed.show(5)
```

### **Pipeline de vectorisation et classification pour l'analyse de sentiment**

> Rappelons que notre but est ici de classer les sentiments en deux catégories : classe 0 pour dire que c'est un sentiment négatif et classe 1 (ici 4) pour dire que c'est un sentiment positif. Notre problème contient deux sous problèmes : le permier consiste à bien vectoriser nos données, et le deuxième à faire la classification. Néanmoins le problème de classification est en soit un probème déjà résolu car nous avons des modèles robustes pour avoir des perfomances très élévées. Le vrai problème est donc de bien vectoriser nos données,car sans cela le modèle de classification pourrait mal apprendre et donc mal généraliser.


> Pour mieux organiser notre travail, nous allons créer des classes de vectorisation qui intégrent directement la classification. Chaque classe utilisera une technique de vectorisation différente sur les mêmes algorithmes de classification.

#### **Étapes du Pipeline**

##### Vectorisation

> Chaque méthode de vectorisation a ses avantages et incovénients, le choix de la technique dépends de la problématique et de la personne qui la developpe, certaines peuvent performer mieux que d'autres dans certains cas. Dans ce projet nous utiliserons 3 méthodes de vectorisation statistiques et une méthode basée sur les réseaux de neurones.


- **TFIDF (Term Frequency-Inverse Document Frequency)** : Mesure l'importance d'un mot dans un document par rapport à l'ensemble des documents, en tenant compte de la fréquence du mot dans le document et son inverse de fréquence dans tous les documents.

- **CountVectorizer** : Convertit le texte en un vecteur de fréquence de termes, où chaque terme est représenté par sa fréquence d'apparition dans le document.

- **HashingTF** : Utilise une fonction de hachage pour convertir les termes en indices et compte leur fréquence, ce qui est utile pour traiter de grands ensembles de données.

- **Word2Vec** : Emploie un modèle neuronal pour mapper les mots dans un espace vectoriel où les mots de significations similaires sont proches les uns des autres.

##### Classification

> Après la vectorisation, nous appliquerons différents modèles de classification pour déterminer le sentiment des textes :

- **Logistic Regression (LR)** : Un modèle simple mais puissant pour les tâches de classification binaire, utilisant une fonction logistique pour modéliser la probabilité de défaut.

- **Support Vector Classifier (SVC)** : Construit un ou plusieurs hyperplans dans un espace de grande dimension pour classer les données.

- **Random Forest (RF)** : Un ensemble d'arbres de décision utilisés pour la classification, offrant une robustesse et une réduction de l'overfitting.

> Ce piple intégré nous permettra de

- Tester et comparer directement l'efficacité de différentes combinaisons de vectorisation et de classification.

- Optimiser les performances du modèle pour une prédiction des sentiments, en utilisant la configuration la plus efficace détectée lors des expérimentations.

### **TFIDF vectorisation & Trainer**

> Nous partons à l'application "numérique" de la description de notre pipline ci-haut. Notre première classe s'appelle : **`TFIDFModelTrainer`** . Elle va utiliser le TFIDF pour vectoriser et les méthodes de classification suivantes : logistic regression, SVC et random forest.

```python
# Définir la classe TFIDFModelTrainer
class TFIDFModelTrainer:
    def __init__(self, max_features=20000):
        self.max_features = max_features
        self.lr_pipeline = None
        self.svc_pipeline = None
        self.rf_pipeline = None
        self.lr_model = None
        self.svc_model = None
        self.rf_model = None
        self.build_pipeline()

    def build_pipeline(self):
        # Vectorisation TF-IDF
        count_vectorizer = CountVectorizer(inputCol="processed_message", outputCol="cv_features", vocabSize=self.max_features)
        idf = IDF(inputCol="cv_features", outputCol="tfidf_features")

        # StandardScaler pour standardiser les caractéristiques
        scaler = StandardScaler(inputCol="tfidf_features", outputCol="scaled_features")

        # Modèles
        lr = LogisticRegression(featuresCol="scaled_features", labelCol="polarity", predictionCol="lr_prediction", regParam=0.5, elasticNetParam=0.0)
        svc = LinearSVC(featuresCol="scaled_features", labelCol="polarity", predictionCol="svc_prediction")
        rf = RandomForestClassifier(featuresCol="scaled_features", labelCol="polarity", predictionCol="rf_prediction", numTrees=10)

        # Pipelines pour chaque modèle
        self.lr_pipeline = Pipeline(stages=[count_vectorizer, idf, scaler, lr])
        self.svc_pipeline = Pipeline(stages=[count_vectorizer, idf, scaler, svc])
        self.rf_pipeline = Pipeline(stages=[count_vectorizer, idf, scaler, rf])

    def fit(self, train_df: DataFrame):
        # Convertir la colonne 'polarity' en entier et binariser les labels
        train_df = train_df.withColumn("polarity", when(col("polarity") == 4, 1).otherwise(0).cast("int"))

        if self.lr_pipeline is None or self.svc_pipeline is None or self.rf_pipeline is None:
            self.build_pipeline()
        self.lr_model = self.lr_pipeline.fit(train_df)
        self.svc_model = self.svc_pipeline.fit(train_df)
        self.rf_model = self.rf_pipeline.fit(train_df)

    def evaluate(self, df: DataFrame):
        df = df.withColumn("polarity", when(col("polarity") == 4, 1).otherwise(0).cast("int"))

        if self.lr_model is None or self.svc_model is None or self.rf_model is None:
            raise ValueError("Models have not been trained yet. Please call fit() first.")

        # Prédictions
        lr_predictions = self.lr_model.transform(df)
        svc_predictions = self.svc_model.transform(df)
        rf_predictions = self.rf_model.transform(df)

        # Évaluateur
        evaluator = MulticlassClassificationEvaluator(labelCol="polarity", predictionCol="prediction", metricName="accuracy")

        # Calcul des précisions
        lr_accuracy = evaluator.evaluate(lr_predictions.withColumnRenamed("lr_prediction", "prediction"))
        svc_accuracy = evaluator.evaluate(svc_predictions.withColumnRenamed("svc_prediction", "prediction"))
        rf_accuracy = evaluator.evaluate(rf_predictions.withColumnRenamed("rf_prediction", "prediction"))

        return {
            "lr_accuracy": lr_accuracy,
            "svc_accuracy": svc_accuracy,
            "rf_accuracy": rf_accuracy
        }

    def transform(self, df: DataFrame):
        if self.lr_model is None or self.svc_model is None or self.rf_model is None:
            raise ValueError("Models have not been trained yet. Please call fit() first.")
        lr_predictions = self.lr_model.transform(df)
        svc_predictions = self.svc_model.transform(df)
        rf_predictions = self.rf_model.transform(df)
        return lr_predictions, svc_predictions, rf_predictions
```

## HashingVectorisation & Trainer

> Dans la suite de notre analyse nous utilisons maintenant le **Hashinhvectoriser** de mlib de spark pour vectoriser et conservons les mêmes algos pour la classification. Contrairement à CountVectorizer, qui crée un vocabulaire de tous les jetons uniques dans les données textuelles, HashingVectorizer n'enregistre pas de vocabulaire. Au lieu de cela, il applique une fonction de hachage à chaque jeton pour le mapper à un index dans un vecteur de sortie de taille fixe. Cela permet à HashingVectorizer d'être plus efficace en termes de mémoire et de temps de traitement, car il n'a pas besoin de stocker un grand vocabulaire en mémoire.

> Cependant, une limitation de HashingVectorizer est qu'il peut y avoir des collisions de hachage, où deux jetons différents sont mappés au même index dans le vecteur de sortie. Cela peut entraîner une perte d'information et une diminution des performances de classification.

```python
# Définir la classe HashingModelTrainer
class HashingModelTrainer:
    def __init__(self, num_features=5000):
        self.num_features = num_features
        self.lr_pipeline = None
        self.svc_pipeline = None
        self.rf_pipeline = None
        self.lr_model = None
        self.svc_model = None
        self.rf_model = None
        self.build_pipeline()

    def build_pipeline(self):
        # Vectorisation HashingTF
        hashing_tf = HashingTF(inputCol="processed_message", outputCol="hashing_features", numFeatures=self.num_features)

        # StandardScaler pour standardiser les caractéristiques
        scaler = StandardScaler(inputCol="hashing_features", outputCol="scaled_features")

        # Modèles
        lr = LogisticRegression(featuresCol="scaled_features", labelCol="polarity", predictionCol="lr_prediction", regParam=0.5, elasticNetParam=0.0)
        svc = LinearSVC(featuresCol="scaled_features", labelCol="polarity", predictionCol="svc_prediction")
        rf = RandomForestClassifier(featuresCol="scaled_features", labelCol="polarity", predictionCol="rf_prediction", numTrees=10)

        # Pipelines pour chaque modèle
        self.lr_pipeline = Pipeline(stages=[hashing_tf, scaler, lr])
        self.svc_pipeline = Pipeline(stages=[hashing_tf, scaler, svc])
        self.rf_pipeline = Pipeline(stages=[hashing_tf, scaler, rf])

    def fit(self, train_df: DataFrame):
        # Convertir la colonne 'polarity' en entier et binariser les labels
        train_df = train_df.withColumn("polarity", when(col("polarity") == 4, 1).otherwise(0).cast("int"))

        self.lr_model = self.lr_pipeline.fit(train_df)
        self.svc_model = self.svc_pipeline.fit(train_df)
        self.rf_model = self.rf_pipeline.fit(train_df)

    def evaluate(self, df: DataFrame):
        df = df.withColumn("polarity", when(col("polarity") == 4, 1).otherwise(0).cast("int"))

        if self.lr_model is None or self.svc_model is None or self.rf_model is None:
            raise ValueError("Models have not been trained yet. Please call fit() first.")

        # Prédictions
        lr_predictions = self.lr_model.transform(df)
        svc_predictions = self.svc_model.transform(df)
        rf_predictions = self.rf_model.transform(df)

        # Évaluateur
        evaluator = MulticlassClassificationEvaluator(labelCol="polarity", predictionCol="prediction", metricName="accuracy")

        # Calcul des précisions
        lr_accuracy = evaluator.evaluate(lr_predictions.withColumnRenamed("lr_prediction", "prediction"))
        svc_accuracy = evaluator.evaluate(svc_predictions.withColumnRenamed("svc_prediction", "prediction"))
        rf_accuracy = evaluator.evaluate(rf_predictions.withColumnRenamed("rf_prediction", "prediction"))

        return {
            "lr_accuracy": lr_accuracy,
            "svc_accuracy": svc_accuracy,
            "rf_accuracy": rf_accuracy
        }

    def transform(self, df: DataFrame):
        if self.lr_model is None or self.svc_model is None or self.rf_model is None:
            raise ValueError("Models have not been trained yet. Please call fit() first.")
        lr_predictions = self.lr_model.transform(df)
        svc_predictions = self.svc_model.transform(df)
        rf_predictions = self.rf_model.transform(df)
        return lr_predictions, svc_predictions, rf_predictions
```

## **Cross Validation et Trainer**

> Cette dernière étape utilise la même configuration que les modèles précédentes, nous utilisons la même logique de vectorisation et de classification. Le petit plus est la recherche d'hyper paramètre avec un `GridSearch CV`. Comme nous l'avons défini au début nous souhaitons tester plusieurs hyper paramètres sur nos modèles en espérant trouver la meilleure combinaison qui nous assure une meilleure performance. Ceci ne garantit pas qu'on trouvera un meilleur modèle car tout dépends du nombre de paramètre que nous mettons dans la grille de recherche. (A noter que l'entrainement peut être super long)

> Mais la logique reste la même, nous avons une classe qui a les mêmes méthodes. Elle prends en paramètre une méthode de vectorisation et de classification et renvoie les meilleures hyper paramètres qui renvoient le meilleur resultat.

```python
from pyspark.sql import SparkSession
from pyspark.ml import Pipeline
from pyspark.ml.feature import HashingTF, CountVectorizer, IDF, StandardScaler
from pyspark.ml.classification import LogisticRegression, LinearSVC
from pyspark.ml.tuning import CrossValidator as MLlibCrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.sql.functions import col, when

class ModelTrainerCV:
    def __init__(self, vectorizer_type, model_type):
        self.vectorizer_type = vectorizer_type
        self.model_type = model_type
        self.cv_model = None

    def build_pipeline(self):
        stages = []
        vectorizer = None
        if self.vectorizer_type == 'hashing':
            vectorizer = HashingTF(inputCol="processed_message", outputCol="features")
            stages = [vectorizer]
        elif self.vectorizer_type == 'count':
            vectorizer = CountVectorizer(inputCol="processed_message", outputCol="raw_features")
            idf = IDF(inputCol="raw_features", outputCol="features")
            stages = [vectorizer, idf]
        elif self.vectorizer_type == 'tfidf':
            count_vectorizer = CountVectorizer(inputCol="processed_message", outputCol="raw_features")
            idf = IDF(inputCol="raw_features", outputCol="features")
            stages = [count_vectorizer, idf]

        scaler = StandardScaler(inputCol="features", outputCol="scaled_features")

        if self.model_type == 'lr':
            model = LogisticRegression(featuresCol="scaled_features", labelCol="polarity", rawPredictionCol="lr_rawPrediction", predictionCol="lr_prediction", probabilityCol="lr_probability")
        elif self.model_type == 'svc':
            model = LinearSVC(featuresCol="scaled_features", labelCol="polarity", rawPredictionCol="svc_rawPrediction", predictionCol="svc_prediction")

        stages += [scaler, model]
        self.pipeline = Pipeline(stages=stages)

        return self.pipeline

    def fit_and_evaluate(self, train_df, test_df):
        train_df = train_df.withColumn("polarity", when(col("polarity") == 4, 1).otherwise(0).cast("int"))
        test_df = test_df.withColumn("polarity", when(col("polarity") == 4, 1).otherwise(0).cast("int"))

        pipeline = self.build_pipeline()

        param_grid = ParamGridBuilder()
        if self.vectorizer_type in ['hashing']:
            param_grid = param_grid.addGrid(pipeline.getStages()[0].numFeatures, [1000, 3000, 5000, 6000, 8000, 20000])
        if self.vectorizer_type in ['count', 'tfidf']:
            param_grid = param_grid.addGrid(pipeline.getStages()[0].vocabSize, [1000, 6000, 2000, 3000])

        if self.model_type == 'lr':
            param_grid = param_grid.addGrid(pipeline.getStages()[-1].regParam, [0.2, 0.5, 0.6]) \
                                   .addGrid(pipeline.getStages()[-1].elasticNetParam, [0.0, 0.2, 0.5])
        elif self.model_type == 'svc':
            param_grid = param_grid.addGrid(pipeline.getStages()[-1].regParam, [0.1, 0.01, 0])

        param_grid = param_grid.build()

        crossval = MLlibCrossValidator(estimator=pipeline,
                                       estimatorParamMaps=param_grid,
                                       evaluator=MulticlassClassificationEvaluator(labelCol="polarity", predictionCol=self.model_type + "_prediction", metricName="accuracy"),
                                       numFolds=5)

        cv_model = crossval.fit(train_df)

        train_predictions = cv_model.transform(train_df)
        test_predictions = cv_model.transform(test_df)

        evaluator = MulticlassClassificationEvaluator(labelCol="polarity", predictionCol=self.model_type + "_prediction", metricName="accuracy")

        train_accuracy = evaluator.evaluate(train_predictions)
        test_accuracy = evaluator.evaluate(test_predictions)

        best_model = cv_model.bestModel
        best_params = {param.name: best_model.stages[-1].getOrDefault(param) for param in param_grid[0] if param.parent == best_model.stages[-1].uid}

        self.cv_model = best_model

        return {
            "train_accuracy": train_accuracy,
            "test_accuracy": test_accuracy,
            "best_params": best_params
        }

    def predict(self, noclass_df):
        predictions = self.cv_model.transform(noclass_df)
        return predictions.select('message', 'processed_message', self.model_type + '_prediction')
```

> Nous avons opter de mettre en validation croisée les deux méthodes de vectorisation et de classification qui ont donné les meilleures performances avec les paramètre par défaut. A savoir respectivement le **TFIDF** et le **CountVectoriser**  avec la **regression logistique** et le **SVC**. Nous essayerons de trouver la meilleure combinaison qui nous trouve les meilleures hyper paramètres.

> Rappelons néanmoins que ça n'implique pas forcément que nous auront un meilleur accuracy que nos modèles précédentes, mais on aurait une meilleure généralisation et éviter l'overfitting. En effet étant donné que c'est nous qui initialisons nos paramètres de GridSearch, la performance dépendra de cela et le temps de calcul peut être surper long.

> Comme le savons pyspark utilise le principe du Map reduce qui est un modèle de programmation créé chez Google pour automatiser le calcul parallèle sur de grandes quantités de données, notamment pour le calcul du PageRank. Conçu chez google Google, il repose sur une stratégie de parallélisation commune observée dans les traitements massivement parallèles nécessaires pour leur moteur de recherche. MapReduce divise les données en sous-ensembles traités en parallèle par différentes machines d'un cluster, puis agrège les résultats intermédiaires pour obtenir le résultat final. Le processus comprend deux principales étapes : "MAP", où une fonction est appliquée à tous les éléments, et "REDUCE", où une fonction agrège les résultats intermédiaires. Les données sont représentées sous forme de paires clé-valeur, permettant de modéliser divers problèmes. Les étapes incluent la division des données, l'application de l'opération map, le tri et le regroupement des résultats intermédiaires, suivis de l'application de l'opération reduce pour obtenir le résultat final. Spark offre des opérations supplémentaires pour le calcul distribué et utilise diverses techniques d'optimisation.



> **Chaque job listé ci-dessus** est lié à l'opération treeAggregate dans RDDLossFunction.scala. Cette méthode est généralement utilisée pour effectuer des agrégations de manière incrémentielle, ce qui est utile pour réduire la communication entre les noeuds dans un environnement distribué comme Spark.

> Un RDD (Resilient Distributed Dataset) est une structure de données  de Spark permettant d'effectuer les calculs distribués. Les RDD offrent deux types de méthodes :

- Les transformations, qui produisent de nouveaux RDD.

- Les actions, qui produisent des résultats finaux non RDD.

> Les transformations sont évaluées paresseusement ("lazy evaluation") uniquement lorsque leur résultat est requis par une action. Cela signifie que les appels à des transformations comme flatMap, map, et reduceByKey sont rapides car ils ne déclenchent pas immédiatement le calcul. Les fonctions passées en argument à ces transformations ne sont exécutées qu'au moment où une action, telle que collect, est appelée.


> Noms des RDDs : Les descriptions des RDDs mentionnent l'application de filtres et de projections sur les données. Ces opérations font partie de transformations qui préparent nos données pour l'analyse de sentiment.
  
> Opérations: C'est l'opératon de CrossValidation qui est traitée.  En effet nous l'avons utilisé dans la denrière partie pour optimiser nos calculs.


> Niveau de stockage : Les RDDs sont stockés en mémoire (RAM) et sur disque de façon désérialisée et sont répliquées une fois. La réplication est une stratégie pour la tolérance aux pannes, garantissant que les données sont disponibles même en cas de défaillance d'un nœud.
    
> Partitions : Chaque **RDD** est divisé en **4 partitions**. Ceci est important pour la distribution et le parallélisme, permettant à Spark de traiter les données sur plusieurs nœuds du cluster.

> Fraction mise en cache : Les RDDs sont entièrement mis en cache (100 %), ce qui indique que toutes les partitions sont en mémoire et prêtes à être accédées rapidement sans rechargement à partir du disque.

> Taille en mémoire : Les tailles en mémoire des RDDs sont de 24,4 MiB et 6,0 MiB. Cela montre que les données nécessaires pour nos analyses sont conservées en mémoire pour un accès rapide.

> Un exécuteur (executor) dans Spark est une unité de travail distribuée responsable de l'exécution des tâches sur un cluster. Chaque exécuteur :

- Gère la mémoire et le stockage local pour les données RDD et les calculs intermédiaires.

- Communique avec le pilote pour envoyer les résultats des tâches et recevoir de nouvelles tâches.
